## Leitura e limpeza dos dados

A célula abaixo puramente lê o arquivo .csv fornecido para um **pandas.DataFrame** e imprime seu cabeçalho e informações.

```
teachers = pd.read_csv(file_name, sep=';')
display(teachers.head())
print("");
display(teachers.info())
```

## Limpeza das linhas

```
graduate = teachers.dropna(subset=['GRADUACAO_LOCAL'])
gradDropCols = teachers.columns[~teachers.columns.str.contains('GRADUACAO')]
gradDropCols = [c for c in gradDropCols if c != 'nome']
graduate = graduate.drop(columns=gradDropCols)
display(graduate.info())
print("")

specialization = teachers.dropna(subset=['ESPECIALIZACAO_LOCAL'])
specDropCols = teachers.columns[~teachers.columns.str.contains('ESPECIALIZACA')]
specDropCols = [c for c in specDropCols if c != 'nome']
specialization = specialization.drop(columns=specDropCols)
display(specialization.info())
print("")

masters = teachers.dropna(subset=['MESTRADO_LOCAL'])
mastDropCols = teachers.columns[~teachers.columns.str.contains('MESTRADO')]
mastDropCols = [c for c in mastDropCols if c != 'nome']
masters = masters.drop(columns=mastDropCols)
display(masters.info())
print("")

doctorate = teachers.dropna(subset=['DOUTORADO_LOCAL'])
doctDropCols = teachers.columns[~teachers.columns.str.contains('DOUTORADO') | teachers.columns.str.contains('PÓS-DOUTORADO')]
doctDropCols = [c for c in doctDropCols if c != 'nome']
doctorate = doctorate.drop(columns=doctDropCols)
display(doctorate.info())
print("")

post_doctoral = teachers.dropna(subset=['PÓS-DOUTORADO_LOCAL'])
postDropCols = teachers.columns[~teachers.columns.str.contains('PÓS-DOUTORADO')]
postDropCols = [c for c in postDropCols if c != 'nome']
post_doctoral = post_doctoral.drop(columns=postDropCols)
display(post_doctoral.info())
print("")
```

A célula de código abaixo gera um dicionário com todos os estados e sua forma geográfica, obtidos do arquivo *GeoJSON*. E, em seguida, itera pelo *DataFrame* dos tweets, para que estes sejam contabilizados no seu estado federativo (basicamente checa se o ponto geográfico está dentro do território do estado, e adiciona seu valor à estatística).
