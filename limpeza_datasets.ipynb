## Limpeza dos dados

No dataset original cada modalidade de ensino vem concatenado:

[tipo] : AnoInicio - AnoFim:CursoUniversidade

Na célula abaixo, quebramos cada uma modalidade em 3 colunas:

**MODALIDADE_ANO, MODALIDADE_CURSO, MODALIDADE_LOCAL**

como a coluna tipo perde seu sentido informativo, a mesma é dropada do dataSet.



```
import pandas as pd

df = pd.read_csv('convertcsv.csv')

df['GRADUACAO_ANO'] =  df['GRADUAÇÃO'].str.split(':').str[0]
df['GRADUACAO_CURSO'] =  df['GRADUAÇÃO'].str.split(':').str[1]
df['GRADUACAO_LOCAL']= "Universidade " + df['GRADUACAO_CURSO'].str.split('Universidade', n=1).str[1]
df['GRADUACAO_CURSO']= df['GRADUACAO_CURSO'].str.split('Universidade', n=1).str[0]
df = df.drop(columns=['GRADUAÇÃO'])

df['ESPECIALIZACAO_ANO'] =  df['ESPECIALIZAÇÃO'].str.split(':').str[0]
df['ESPECIALIZACAO_CURSO'] =  df['ESPECIALIZAÇÃO'].str.split(':').str[1]
df['ESPECIALIZACAO_LOCAL']= "Universidade " + df['ESPECIALIZACAO_CURSO'].str.split('Universidade', n=1).str[1]
df['ESPECIALIZACAO_CURSO']= df['ESPECIALIZACAO_CURSO'].str.split('Universidade', n=1).str[0]
df = df.drop(columns=['ESPECIALIZAÇÃO'])

df['MESTRADO_ANO'] =  df['MESTRADO'].str.split(':').str[0]
df['MESTRADO_CURSO'] =  df['MESTRADO'].str.split(':').str[1]
df['MESTRADO_LOCAL']= "Universidade " + df['MESTRADO_CURSO'].str.split('Universidade', n=1).str[1]
df['MESTRADO_CURSO']= df['MESTRADO_CURSO'].str.split('Universidade', n=1).str[0]
df = df.drop(columns=['MESTRADO'])

df['DOUTORADO_ANO'] =  df['DOUTORADO'].str.split(':').str[0]
df['DOUTORADO_CURSO'] =  df['DOUTORADO'].str.split(':').str[1]
df['DOUTORADO_LOCAL']= "Universidade " + df['DOUTORADO_CURSO'].str.split('Universidade', n=1).str[1]
df['DOUTORADO_CURSO']= df['DOUTORADO_CURSO'].str.split('Universidade', n=1).str[0]
df = df.drop(columns=['DOUTORADO'])


df['PÓS-DOUTORADO_ANO'] =  df['PÓS-DOUTORADO'].str.split(':').str[0]
df['PÓS-DOUTORADO_CURSO'] =  df['PÓS-DOUTORADO'].str.split(':').str[1]
df['PÓS-DOUTORADO_LOCAL']= "Universidade " + df['PÓS-DOUTORADO_CURSO'].str.split('Universidade', n=1).str[1]
df['PÓS-DOUTORADO_CURSO']= df['PÓS-DOUTORADO_CURSO'].str.split('Universidade', n=1).str[0]
df = df.drop(columns=['PÓS-DOUTORADO'])




#df = df.fillna('NULL')
df.head()

```

## Inserindo Geolocalização

Usando geocoder, neste ponto inserimos a latitude e longitude por docente de cada Universidade que fez um curso.


```
import geocoder
from tqdm import tqdm

for i in tqdm(range(len(df))):
  if df.loc[i, 'GRADUACAO_LOCAL'] != "NULL":
    g = geocoder.arcgis(df.loc[i, 'GRADUACAO_LOCAL'])
    df.loc[i,'GRADUACAO_Latitude'] = g.lat
    df.loc[i,'GRADUACAO_Longitude'] = g.lng

  if df.loc[i, 'ESPECIALIZACAO_LOCAL'] != 'NULL':
    g = geocoder.arcgis(df.loc[i, 'ESPECIALIZACAO_LOCAL'])
    df.loc[i,'ESPECIALIZACAO_Latitude'] = g.lat
    df.loc[i,'ESPECIALIZACAO_Longitude'] = g.lng
  
  if df.loc[i, 'MESTRADO_LOCAL'] != 'NULL':
    g = geocoder.arcgis(df.loc[i, 'MESTRADO_LOCAL'])
    df.loc[i,'MESTRADO_Latitude'] = g.lat
    df.loc[i,'MESTRADO_Longitude'] = g.lng
  
  if df.loc[i, 'DOUTORADO_LOCAL'] != "NULL":
    g = geocoder.arcgis(df.loc[i, 'DOUTORADO_LOCAL'])
    df.loc[i,'DOUTORADO_Latitude'] = g.lat
    df.loc[i,'DOUTORADO_Longitude'] = g.lng
  
  if df.loc[i, 'PÓS-DOUTORADO_LOCAL'] != 'NULL':
    g = geocoder.arcgis(df.loc[i, 'PÓS-DOUTORADO_LOCAL'])
    df.loc[i,'PÓS-DOUTORADO_Latitude'] = g.lat
    df.loc[i,'PÓS-DOUTORADO_Longitude'] = g.lng
  
df.head()
```
